```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from math import sqrt

# Load the simulator
sim = Aer.get_backend('aer_simulator')

# Create quantum circuit and initialize to |1‚ü©
qc = QuantumCircuit(1)
initial_state = [0, 1]
qc.initialize(initial_state, 0)
qc.save_statevector()

# Run simulation
result = sim.run(qc).result()
statevector = result.get_statevector()

print("Statevector:", statevector)
```

```python
from qiskit.visualization import plot_histogram

# Add measurement to the same circuit
qc.measure_all()

# Run the updated circuit with measurement
result = sim.run(qc).result()
counts = result.get_counts()

# Display measurement result as a histogram
plot_histogram(counts)
```

```python
# Define a new quantum circuit for a complex initial state
qc2 = QuantumCircuit(1)
custom_state = [1/sqrt(3), 1j/sqrt(3/2)]
qc2.initialize(custom_state, 0)
qc2.save_statevector()

# Run simulation
result2 = sim.run(qc2).result()
custom_vector = result2.get_statevector()

print("Custom Statevector:", custom_vector)
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
from math import sqrt
from qiskit.quantum_info import Statevector

# Setup simulator
sim = Aer.get_backend('aer_simulator')

# Define custom state
initial_state = [1/sqrt(3), 1j/sqrt(3/2)]

# Create a new quantum circuit
qc = QuantumCircuit(1, 1)
qc.initialize(initial_state, 0)

# Add measurement
qc.measure(0, 0)

# Run simulation
result = sim.run(qc, shots=1024).result()
counts = result.get_counts()

# Display histogram
plot_histogram(counts)
```
